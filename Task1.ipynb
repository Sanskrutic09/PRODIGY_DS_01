## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Dataset

```python
data=pd.read_csv('world_population.csv')
```

# Displaying First Few Rows

```python
data.head()
```

# Get Dimensions of Dataframe

```python
data.shape
```

# Summary statistics of numeric columns


```python
data.describe()
```

# Sort by population and get top 5 Capitals

```python
top_capitals = data.sort_values(by="World Population Percentage", ascending=False).head(5)
```

# Plot the top 5 populous Capitals

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Capital", y="2022 Population", data=top_capitals,color='green')
plt.title("Top 5 Most Populous Capitals in 2022")
plt.xlabel("Capital",)
plt.ylabel("2022 Population")
plt.tight_layout()
plt.show()
```

 # Group data by continent and sum population for 2022

```python
continent_population = data.groupby("Continent")["2022 Population"].sum().reset_index()
```

# Plot population distribution by continent in 2022

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Continent", y="2022 Population", data=continent_population,color='green')
plt.title("Population Distribution by Continent in 2022")
plt.xlabel("Continent")
plt.ylabel("Total Population in 2022")
plt.tight_layout()
plt.show()
```

# Calculate population growth from 2010 to 2022

```python
data["Growth (2010 to 2022)"] = data["2022 Population"] - data["2010 Population"]
```

# Sort by the population growth from 2010 to 2022

```python
top_growth_countries = data.sort_values(by="Growth (2010 to 2022)", ascending=False).head(5)

plt.figure(figsize=(8, 5))
sns.barplot(x="Country/Territory", y="Growth (2010 to 2022)", data=top_growth_countries,color='green')
plt.title("Top 5 Countries by Population Growth from 2010 to 2022")
plt.xlabel("Country")
plt.ylabel("Population Growth (2010 to 2022)")
plt.tight_layout()
plt.show()
```

# Sort by population density in 2022 and get the top  5 countries

```python
top_density = data.sort_values(by="Density (per km²)", ascending=False).head(5)

plt.figure(figsize=(8, 5))
sns.barplot(x="Country/Territory", y="Density (per km²)", data=top_density,color='green')
plt.title("Top 5 Countries by Population Density in 2022")
plt.xlabel("Country")
plt.ylabel("Density (per km²)")

plt.tight_layout()
plt.show()
```

# Plot Area vs. Population 2022

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Area (km²)", y="2022 Population", data=data, color="blue")
plt.title("Area vs. Population in 2022")
plt.xlabel("Area (km²)")
plt.ylabel("2022 Population")
plt.tight_layout()
plt.show()
```
